```python
import pandas as pd
```

```python
x = pd.read_excel('cars.xlsx')
```

```python
x
```

```python
df=x.copy()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
imputer=imputer.fit(df[['hwy']])
df['hwy'] = imputer.transform(df[['hwy']]).ravel()
```

```python
df['hwy'].mean()
```

```python
df['displ']=df['displ'].fillna(method='backfill')
```

```python
df['cty']=df['cty'].fillna(method='backfill')
```

```python
df.isnull().sum()
```

```python
df.corr() #cyl and displ have highest correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['hwy'],df['cty']) #there is a linear relationship between cty and hwy variables which indicates that there's a correlation between car's fuel economy in the city and highway
```

```python
sns.pairplot(df,hue = 'drv',palette = 'coolwarm') #there's a strong negative correlation between displ and hwy variables
```

```python
sns.countplot(x='class',data=df)
```

```python
sns.barplot(x='drv',y='cyl',data=df) #reverse drivetrain is most frequent
```

```python
sns.boxplot(x="displ", y="class", data=df,palette='rainbow')
```

```python
sns.violinplot(x="displ", y="class", data=df,palette='rainbow')
```
